```python
imdb = pd.read_csv('imdb.csv')
```

```python
imdb.head()
```

```python
imdb.info()
```

```python
print("Average rating: ",imdb['Rating'].mean())
print("Rating of 99% movies lies below: ",imdb['Rating'].quantile(0.99))
print("Rating of 80% movies lies below: ",imdb['Rating'].quantile(0.95))
print("Rating of 50% movies lies below: ",imdb['Rating'].quantile(0.50))
print("Maximum rating of a movie: ",imdb['Rating'].max())
```

```python
clean = imdb.dropna(axis=0)
clean.shape
clean = clean.rename(columns = {'Revenue (Millions)' : 'Revenue','Runtime (Minutes)' : 'Runtime'})
```

```python
s = clean['Genre'].str.split(',').apply(Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = 'Genre'
del clean['Genre']
split_genres = clean.join(s)
```

```python
#data.loc[ data['Rating'] > data['Rating'].quantile(0.95)] = '8.1+'
plt.figure(figsize=(20,12))
sbn.countplot(clean['Rating'].astype('str').sort_values())
plt.ylabel("Movies Count")
plt.show()
```

```python
print("Maximum revenue of a movie: ",clean['Revenue'].max())
print("Maximum revenue of a movie: ",clean['Revenue'].min())
print("Average revenue of all movies: ",clean['Revenue'].mean())
```

```python
clean.describe()
```

# Corelation Between Revenue and Rating

```python
sbn.jointplot(x='Rating',y='Revenue', data=clean, height=10, ratio=3, color='r')
plt.show()
```

### Runtime and Revenue Corelation

```python
sbn.jointplot(x='Runtime',y='Revenue', data=clean, height=10, ratio=3, color='r')
plt.show()
```

## Correlation between Revenue and Rating

```python
rating = clean.copy()
rating['RatingCategory'] = pd.cut(rating['Rating'],[-1,0,3,5,7,8,9], labels=['0 Rating','1-3 Rating','4-5 Rating','6-7 Rating','8+ Rating','9+ rating'])
plt.figure(figsize=(15,10))
sbn.boxplot(x='RatingCategory', y='Revenue', data=rating)
plt.show()
```

# Histogram

```python
clean.hist(color='r',figsize=(15,15))
plt.show()
```

```python
#Top 10 Movies with highest revenue 
topr = clean.nlargest(10,'Revenue')
```

# Top 10 movies by revenue

```python
plt.figure(figsize=(13,8))

plt.bar(topr['Title'],topr['Revenue'])
plt.xticks(rotation=90)
plt.xlabel("Movies",fontsize="15")
plt.ylabel('Revenue',fontsize="15")
plt.title("Top 10 movies by revenue",fontsize="20")
plt.show()
```

```python
plt.figure(figsize=(15,8))

plt.scatter(topr['Title'],topr['Revenue'])
plt.xticks(rotation=90)
plt.xlabel("Movies",fontsize="15")
plt.ylabel('Revenue',fontsize="15")
plt.title("Top 10 movies by revenue",fontsize="20")
plt.show()
```

# Top 10 movies by rating

```python
#Top 10 Movies with highest rating
toprat = clean.nlargest(10,'Rating')
```

```python
plt.figure(figsize=(18,10))

plt.bar(toprat['Title'],toprat['Rating'])
plt.xlabel("Movies", fontsize="15")
plt.ylabel('Rating',fontsize="15")
plt.title("Top 10 movies by rating",fontsize="25")
plt.show()
```

```python
topr = clean.nlargest(10,'Revenue')
plt.figure(figsize=(15,8))

plt.scatter(topr['Title'],topr['Revenue'])
plt.xticks(rotation=90)
plt.xlabel("Movies",fontsize="15")
plt.ylabel('Revenue',fontsize="15")
plt.title("Top 10 movies by revenue",fontsize="20")
plt.show()
```

# Number of movies released in years

```python
year = pd.DataFrame(clean.groupby('Year').Title.nunique())
year.head(15)
```

### In this graph we can see that each year number of movies made is increasing

```python
year.plot.line(title = 'Movies made each year',color='r',figsize=(9,8))
plt.show()
```

```python
genre_list = (pd.DataFrame(split_genres.groupby('Genre').Title.nunique())).sort_values('Title', ascending=True)
```

## Percentage of movies by Genre

```python

genre_list['Title'].plot.pie(title= 'Percentage of movies by genre', figsize=(10,10), autopct='%1.1f%%',fontsize=15,);
```

## Number of movies by Genre

```python
genre_list['Title'].plot.barh(title = 'Movies per Genre',color='r', figsize=(10, 9));
```

```python
genre_mean = split_genres.groupby(['Genre']).mean()
```

## Revenue by Genre

```python
genre_mean[['Revenue']].plot.barh(stacked=False, title = 'Revenue by Genre',color=('DarkBlue','c'), figsize=(15, 10));
```
